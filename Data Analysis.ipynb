```python
#importing lib

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#loading the dataset

df  = pd.read_csv('co-ordinates.csv')
df
```

```python
#Filtering the data

df = df[['date', 'latitude', 'longitude', 'depth', 'mag']]
df
```

```python
#defining the variables for visualizing data

years = pd.DatetimeIndex(df['date']).year
magnitude = df['mag']
depth = df['depth']
```

```python
#number of earthquakes happened every year(1973-2021)

plt.figure(figsize=(25,8))
sns.set(font_scale=1.0)
sns.countplot(x=years)
plt.title("Number of Earthquakes every year")
plt.ylabel('Number of Earthquakes')
plt.xlabel('Year')
plt.show()
```

```python
#Intensity of magnitude over the years

fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
plt.title("Magnitude")
plt.xlabel("Year")
plt.ylabel("Magnitude")
ax.plot(years,magnitude)
plt.show()
```

```python
#plotting the magnitudes over the intensity

#plot1
yr_mag = list(zip(years, magnitude))

mag_lessthan_2 = []
for i,j in yr_mag:
    if j <= 2.0:
        mag_lessthan_2.append([i,j])
        
year_lessthan_2 = []
mags_lessthan_2 = []
for i,j in mag_lessthan_2:
    year_lessthan_2.append(i)
    mags_lessthan_2.append(j)

fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
plt.title("Magnitude < = 2.0")
plt.xlabel("Years")
plt.ylabel("Magnitude")
ax.plot(year_lessthan_2,mags_lessthan_2)

#plot2
mag_greaterthan_2 = []
for i,j in yr_mag:
    if j >= 2.0:
        mag_greaterthan_2.append([i,j])
        
year_greaterthan_2 = []
mags_greaterthan_2 = []

for i,j in mag_greaterthan_2:
    year_greaterthan_2.append(i)
    mags_greaterthan_2.append(j)
    
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
plt.title("Magnitude > = 2.0")
plt.xlabel("Years")
plt.ylabel("Magnitude")
ax.plot(year_greaterthan_2,mags_greaterthan_2)

#plot3
mag_greaterthan_5 = []
for i,j in yr_mag:
    if j >= 5.0:
        mag_greaterthan_5.append([i,j])
        
year_greaterthan_5 = []
mags_greaterthan_5 = []

for i,j in mag_greaterthan_5:
    year_greaterthan_5.append(i)
    mags_greaterthan_5.append(j)
    
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
plt.title("Magnitude > = 5.0")
plt.xlabel("Years")
plt.ylabel("Magnitude")
ax.plot(year_greaterthan_5,mags_greaterthan_5)

#plot4
mag_greaterthan_7 = []
for i,j in yr_mag:
    if j >= 7.0:
        mag_greaterthan_7.append([i,j])
        
year_greaterthan_7 = []
mags_greaterthan_7 = []

for i,j in mag_greaterthan_7:
    year_greaterthan_7.append(i)
    mags_greaterthan_7.append(j)
    
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
plt.title("Magnitude > = 7.0")
plt.xlabel("Years")
plt.ylabel("Magnitude")
ax.plot(year_greaterthan_7,mags_greaterthan_7)
```

```python
plt.figure(figsize=(18,6))
plt.title('Magnitude vs Depth')
plt.xlabel('Magnitude')
plt.ylabel('Depth')
plt.scatter(magnitude,depth)
plt.show()
```
